```python
#dhimas nandista
#G6501211010

import math

class keanggotaan:

    def fuzzy(self, x):
            
            result = {'score':[],'kelas':[]}
        
            if(x <= 0):
                result['score'].append(0)
                result['kelas'].append("Rendah")
                
            if(x > 0 and x <= 25):
                result['score'].append((x-0)/(25-0))
                result['kelas'].append("Rendah")
                
            if(x > 25 and x <= 50):
                result['score'].append((x-25)/(50-25))
                result['kelas'].append("Rendah")
                
            if(x > 50 and x <= 125):
                result['score'].append(1/1-((x-50)/125))
                result['kelas'].append("Sedang")
                
            if(x > 75 and x <= 125):
                result['score'].append(1/1-((x-75)/125))
                result['kelas'].append("Tinggi")
                
            if(x > 75 and x <= 100):
                result['score'].append((x-75)/(100-7))
                result['kelas'].append("Sangat Tinggi")
            
            return result
        
    def operatorAND(self, a, b, c):
        return (min(a, b, c))
    
    def operatorOr(self, a):
        return (max(a))
    
    def operatorNot(self, a):
        return (1-a)
    
    def baik(self, a, b, c):
        x = self.operatorAND(a,b,c)
        return (x*(75-0))
        
    def buruk(self, a, b, c):
        x = self.operatorAND(a,b,c)
        return ((x*(100-50))+50)
    
def main():

    Option = keanggotaan()

    #penilaian karyawan

    score = {'loyalitas':[],'kedisiplinan':[],'prestasi':[]}
    score['loyalitas'] = Option.fuzzy(80)
    score['kedisiplinan'] = Option.fuzzy(32)
    score['prestasi'] = Option.fuzzy(77)
    
  
    #kelas loyalitas
    a = 0
    d = 0
    h = 0
    for i in range (len (score['loyalitas']['kelas'])):
        if(score['loyalitas']['kelas'][i] == 'Tinggi'):
            a = score['loyalitas']['score'][i]
            
        if(score['loyalitas']['kelas'][i] == 'Sedang'):
            d = score['loyalitas']['score'][i]
        
        if(score['loyalitas']['kelas'][i] == 'Sangat Tinggi'):
            h = score['loyalitas']['score'][i]
    b = 0
    l = 0
    for j in range (len (score['kedisiplinan']['kelas'])):
        if(score['kedisiplinan']['kelas'][j] == 'Rendah'):
            b = score['kedisiplinan']['score'][j]
            
        if(score['loyalitas']['kelas'][j] == 'Tinggi'):
            l = score['loyalitas']['score'][j]
    
    c = 0
    f = 0
    m = 0
    for k in range (len (score['prestasi']['kelas'])):
        if(score['prestasi']['kelas'][k] == 'Sangat Tinggi'):
            c = score['prestasi']['score'][k]
            
        if(score['prestasi']['kelas'][k] == 'Tinggi'):
            f = score['prestasi']['score'][k]
        
        if(score['prestasi']['kelas'][k] == 'Sedang'):
            m = score['prestasi']['score'][k]
            
   
    kinerja = {'kelas':[], 'score':[]}
    #deffuzy
    #R1
    kinerja['kelas'].append("Baik")
    kinerja['score'].append(Option.baik(a,b,c))
    
    #R2
    kinerja['kelas'].append("Buruk")
    kinerja['score'].append(Option.buruk(d,b,f))
    
    #R3
    kinerja['kelas'].append("BAIK")
    kinerja['score'].append(Option.baik(h,l,m))
    
    #R4
    kinerja['kelas'].append("Buruk")
    kinerja['score'].append(Option.buruk(a,b,m))
    
    print(kinerja)
            
if __name__ == "__main__":
    main()
    
    
   
```
